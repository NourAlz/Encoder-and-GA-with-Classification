```python
import pandas as pd
from docx import Document
from docx.shared import Pt, Inches, RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.oxml.ns import qn
from docx.oxml import OxmlElement

# Read the Excel file
excel_file = 'SparsePCA_full_grid_results_truncate2d.xlsx'
df = pd.read_excel(excel_file)

# Create a new Word document
doc = Document()

# Add a title (customize as needed)
title = doc.add_paragraph()
title_run = title.add_run("Table 2: Classification using SparsePCA+GA (SVM fitness) → SVM classifier")
title_run.bold = True
title_run.font.size = Pt(12)
title.alignment = WD_ALIGN_PARAGRAPH.CENTER

# Add some space
doc.add_paragraph()

# Create a table with the data (headers + rows)
table = doc.add_table(rows=len(df) + 1, cols=len(df.columns))
table.style = 'Light Grid Accent 1'

# Add header row
header_cells = table.rows[0].cells
for i, column_name in enumerate(df.columns):
    header_cells[i].text = str(column_name)
    # Format header
    for paragraph in header_cells[i].paragraphs:
        for run in paragraph.runs:
            run.font.bold = True
            run.font.size = Pt(11)
        paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

# Add data rows
for row_idx, (idx, row) in enumerate(df.iterrows(), 1):
    row_cells = table.rows[row_idx].cells
    for col_idx, value in enumerate(row):
        row_cells[col_idx].text = str(value)
        # Center align and format cells
        for paragraph in row_cells[col_idx].paragraphs:
            paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
            for run in paragraph.runs:
                run.font.size = Pt(11)

# Set column widths to be even
for row in table.rows:
    for idx, cell in enumerate(row.cells):
        cell.width = Inches(1.2)

# Save the document
output_file = 'SparsePCA_full_grid_results_truncate2d.docx'
doc.save(output_file)

print(f"Word document created successfully: {output_file}")
print(f"Table contains {len(df)} data rows with {len(df.columns)} columns")
```
